```python
import requests, time
import pandas as pd
```

```python
your_key = open('your_key.txt', 'r').read()
```

```python
def articles_df(init, finit):
    ''' retorna um dataframe com informações a respeito de artigos 
        publicados na NYT durante todo o período compreendido entre
        os anos init e finit
        
        ex. init 2020 e finit 2022, retorna dataframe de 2020, 2021 e 2022'''

    years= [j for j in range(init,finit+1)]
    df_articles=[] #lista dos dataframes, cada elemento traz informações sobre um mês-ano

    for year in range(len(years)):
        for month in range(1,13):
            query_url_archive = f"https://api.nytimes.com/svc/archive/v1/{years[year]}/{month}.json?api-key={your_key}"
            archive_request = requests.get(query_url_archive)     
            #d1=archive_request.json()
            #d2=pd.json_normalize(d1['response']['docs'])
            df_articles.append(archive_request)
            time.sleep(6)
    return df_articles

```

```python
def articles_df2(archives):
    ''' retorna um dataframe com informações a respeito de artigos 
        publicados na NYT durante todo o período compreendido entre
        os anos init e finit
        
        ex. init 2020 e finit 2022, retorna dataframe de 2020, 2021 e 2022'''

    df_articles2=[] #lista dos dataframes, cada elemento traz informações sobre um mês-ano

    for i in range(35):
        #query_url_archive = f"https://api.nytimes.com/svc/archive/v1/{years[year]}/{month}.json?api-key={your_key}"
        #archive_request = requests.get(query_url_archive)     
        d1=archives[i].json()
        d2=pd.json_normalize(d1['response']['docs'])
        df_articles2.append(d2)
        print(f'{i}ok')
        
    return df_articles2
```

```python
def sel_columns(dataset):
    
    '''restringe as colunas de interesse para cada um dos meses da lista de datasets'''
    
    
    cleaned_columns = []
    for i in range(len(dataset)):
        dataset[i]=dataset[i][['abstract', 'web_url', 'snippet', 'lead_paragraph', 'print_section',
       'print_page', 'source', 'keywords', 'pub_date',
       'document_type', 'news_desk', 'section_name', 'type_of_material', '_id',
       'word_count', 'uri', 'headline.main', 
       'headline.print_headline', 'byline.original']]
        cleaned_columns.append(dataset[i])
    return cleaned_columns
```

```python
def concat_df (dataset_list):
    '''concatenar datasets presentes em uma lista '''
    limit=len(dataset_list)
    i=1
    while i<= limit:
        dataset_list[i]=pd.concat([dataset_list[i-1], dataset_list[i]], ignore_index=True) 
        i+=1
    df=dataset_list[limit]
    return df
```
